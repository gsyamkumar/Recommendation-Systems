```python
#Importing Libraries

import pandas as pd 
import numpy as np
```

```python
#Loading dataset

df = pd.read_csv("bookRS.csv", encoding='latin1')
df
```

```python
df_all = df.rename(columns = {"User.ID" :"userid", "Book.Title" :"booktitle", "Book.Rating": "rating"})
df_all
```

```python
df_all.sort_values(by=['userid'])
```

```python
#no. of unique users in the dataset
len(df_all.userid.unique())
```

```python
len(df_all.booktitle.unique())
```

```python
df_all.iloc[:,1:]
```

```python
user_df = df_all.pivot_table(index='userid',columns='booktitle',values='rating').reset_index(drop=True)
user_df
```

```python
user_df.index = df_all.userid.unique()
```

```python
user_df
```

```python
#apply imputation
user_df.fillna(0,inplace = True)
```

```python
user_df
```

```python
#calculate cosine similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim = 1 - pairwise_distances(user_df.values,metric='cosine')
```

```python
user_sim
```

```python
#store results
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = df_all.userid.unique()
user_sim_df.columns = df_all.userid.unique()
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df.iloc[0:5,0:5]
```

```python
#most similar users
user_sim_df.idxmax(axis=1)[0:5]
```

```python
#IBCF
#Fetching ratings for Clara Callan
Clara_Callan_rating = user_df['Classical Mythology']
```

```python
Clara_Callan_rating
```

```python
#finding the correlation with different books
similar_to_Clara= user_df.corrwith(Clara_Callan_rating)
```

```python
similar_to_Clara.head(10)
```

```python
corr_Clara = pd.DataFrame(similar_to_Clara,columns=['Correlation'])
corr_Clara.dropna(inplace=True)
corr_Clara.head()
                  
```

```python
corr_Clara[corr_Clara['Correlation'] > 0].sort_values(by='Correlation', ascending = False).head(5)
```
